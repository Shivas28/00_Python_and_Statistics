```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivan', 'Alice'],
    'Age': [25, 30, 35, 40, 22, 28, 45, 32, 29, 25],
    'City': ['NY', 'LA', 'CHI', 'HOU', 'PHX', 'PHL', 'SAT', 'SDG', 'DAL', 'NY'],
    'Salary': [50000, 60000, 70000, np.nan, 55000, 65000, 80000, 72000, 58000, 50000] 
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.drop(columns=['City'])
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df
```
