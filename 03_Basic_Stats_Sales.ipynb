```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("sales_data_with_discounts.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python

df.mode()	
```

```python
df[['Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value','Volume']].mode()
```

```python
numerical_cols=['Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value','Volume']
numerical_cols
```

```python
df[numerical_cols].std()
```

Looking at these numbers, there's big difference in our sales values. Since the average sales number is higher than the middle number (the median), it looks like the data is skewed by a few really big transactions

```python


a = pd.DataFrame({
    'Mean': df[numerical_cols].mean(),
    'Median': df[numerical_cols].median(),
    'Mode': [df[col].mode().iloc[0] for col in numerical_cols],
    'Std Dev': df[numerical_cols].std(),
    'Min': df[numerical_cols].min(),
    'Max': df[numerical_cols].max()
})

print(a.round(2))
```

```python
df[numerical_cols].hist(bins=10,figsize=(12,8))
plt.suptitle("histogram")
plt.show()
```

The histograms show us how the sales are spread out. You can see that most of the sales are for smaller amounts, which means most of our customers are buying lower-priced items

```python
plt.figure(figsize=(12,8))
sns.boxplot(df[numerical_cols])
plt.title("box plot")
plt.xticks(rotation=45)
plt.show()
```

The box plots are really helpful for spotting 'outliers'—basically, the far data points that don't fit the normal pattern. Those dots sitting outside the box are transactions that were way higher or lower than the rest.

```python
categorical_col=["Date", "Day", "SKU", "City", "BU", "Brand", "Model"]
for col in categorical_col:
    plt.figure(figsize=(12,8))
    sns.countplot(x=df[col])
    plt.title("bar chart")
    plt.show()
```

The bar charts give us a quick and easy way to see what's actually selling. This helped me see exactly which brands and business units are moving the most product

Here the stats gave us a solid picture of what's happening with our sales. We found that most sales are on the lower end, we caught some interesting outliers, and we know exactly which categories are the most popular...
