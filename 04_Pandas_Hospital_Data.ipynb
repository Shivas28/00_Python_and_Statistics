```python
import pandas as pd
df_patient = pd.read_csv('Patient_Data.csv')
df_billing = pd.read_csv('Billing_Data.csv')
```

```python

df_patient = df_patient[['PatientID', 'Department', 'Doctor', 'BillAmount']]
```

```python

df_patient = df_patient.drop_duplicates(subset=['PatientID'])
```

```python

mean_bill = df_patient['BillAmount'].mean()
df_patient['BillAmount'] = df_patient['BillAmount'].fillna(mean_bill)
```

```python
dept_revenue = df_patient.groupby('Department')['BillAmount'].sum()
print(dept_revenue)
```

```python
df_merged = pd.merge(df_patient, df_billing, on='PatientID')
```

```python
new_week_patients = pd.DataFrame({
    'PatientID': [2222], 
    'Department': ['Cardio'], 
    'Doctor': ['DR.Samaram'], 
    'BillAmount': [2500]
})
```

```python
df_merged = pd.concat([df_merged, new_week_patients])
```

```python
billing_cats = df_billing[['InsuranceCovered', 'FinalAmount']]
```

```python
df_final = pd.concat([df_merged.reset_index(drop=True), billing_cats])
```
