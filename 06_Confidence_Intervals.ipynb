We need a 99% confidence interval using the sample standard deviation. Since the test destroys the print‑heads, we only have a small sample (n = 15). We don’t know the real population standard deviation, so we use the sample one. And because the sample size is less than 30, we use the t‑distribution instead of the normal z‑distribution. The t‑distribution is made to handle the extra uncertainty that comes with small samples.


```python
import numpy as np
from scipy.stats import t
from scipy.stats import norm
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
n = len(data)
n
```

```python
mean_val = np.mean(data)
sample_sd = np.std(data, ddof=1)
```

```python
print(mean_val)
sample_sd
```

```python
df=n-1
t_crit =t.ppf(0.995, df)
print(t_crit)
```

```python
margin_error_t= t_crit * (sample_sd / np.sqrt(n))
lower_t=mean_val - margin_error_t
upper_t=mean_val + margin_error_t
```

```python
print("99% CI using t-distribution:", lower_t, "to", upper_t)
```

We want a 99% confidence interval, but this time we actually know the population standard deviation (0.2 million characters). Since we know it, we don’t need to estimate it from the sample. That means we can use the z‑distribution instead of the t‑distribution to build the interval.


```python
pop_sd = 0.2
z_crit = norm.ppf(0.995)
print("z-critical value:", z_crit)
```

```python
margin_error_z = z_crit * (pop_sd / np.sqrt(n))
lower_z = mean_val - margin_error_z
upper_z = mean_val + margin_error_z
print("99% CI using z-distribution:", lower_z, "to", upper_z)
```

the company can be 99% sure that an average print‑head will last about 1.1M characters. This lets them set quality guarantees without breaking more units for testing.
