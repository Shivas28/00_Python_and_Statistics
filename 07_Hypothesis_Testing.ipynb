Hypotheses:
Null (H0): The true mean weekly cost is 4000. The formula is still right.


Alternative (H1): The true mean weekly cost is more than 4000. This supports the owners’ claim that costs have gone up.


```python
import math
from scipy.stats import norm

n=25
x_bar =3050
```

```python
mu =1000+(5*600)
sigma=5*25
```

```python
print(mu)
sigma
```

```python
z_calc=(x_bar-mu)/(sigma/math.sqrt(n))
print(z_calc)
```

```python
p_value= 1-norm.cdf(z_calc)
print("p-value==", p_value)
```

```python
z_crit = norm.ppf(0.95)
print("critical value:", z_crit)
```

Since z is not in the rejection region, we fail to reject H0.

There is no evidence that weekly costs are higher than the model predicts. In fact, the sample mean (3050) is much lower than the theoretical mean (4000). The model is not underestimating costs.


Hypotheses:
- H0:mu= 4000
- H1:mu> 4000
